````
# Inteligencia Artificial y Redes Neuronales
## Examen de Medio Curso
### Parte 1
####**Objetivo:** Utilizar las librerías numpy, pandas y matplotlib para procesar, limpiar y extraer información valiosa de un conjunto de datos real sobre el rendimiento y las especificaciones mecánicas de diversos motores y vehículos.
######**Dataset:** Auto MPG
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv"
df = pd.read_csv(url)

print("Primeras 5 filas:")
print(df.head())

print("\nTipos de datos:")
print(df.dtypes)

df["horsepower"] = pd.to_numeric(df["horsepower"], errors="coerce")

print("\nNulos en horsepower antes:")
print(df["horsepower"].isnull().sum())

mediana_hp = np.median(df["horsepower"].dropna())
print("\nMediana de horsepower:", mediana_hp)

df["horsepower"] = df["horsepower"].fillna(mediana_hp)

print("\nNulos en horsepower después:")
print(df["horsepower"].isnull().sum())

df["relacion_peso_potencia"] = df["weight"] / df["horsepower"]

print("\nPrimeras 5 filas con la nueva columna:")
print(df.head())
````

````

# Parte 2: Visualización de Datos (Matplotlib)

# 1. Histograma de la aceleración
# HISTOGRAMA
plt.figure(figsize=(8,5))
plt.hist(df["acceleration"], bins=15, edgecolor="black")
plt.title("Distribución de la aceleración de los vehículos")
plt.xlabel("Aceleración")
plt.ylabel("Frecuencia")
plt.show()

# 2. Gráfico de dispersión: displacement vs horsepower
plt.figure(figsize=(8, 5))
plt.scatter(df["displacement"], df["horsepower"])
plt.title("Desplazamiento del motor vs Caballos de fuerza")
plt.xlabel("Desplazamiento del motor (displacement)")
plt.ylabel("Caballos de fuerza (horsepower)")
plt.show()

# 3. Gráfico de barras: promedio de mpg por país de origen
promedio_mpg_origen = df.groupby("origin")["mpg"].mean()

plt.figure(figsize=(8, 5))
plt.bar(promedio_mpg_origen.index, promedio_mpg_origen.values, edgecolor="black")
plt.title("Promedio de rendimiento (MPG) por país de origen")
plt.xlabel("País de origen")
plt.ylabel("Promedio de MPG")
plt.show()
````
